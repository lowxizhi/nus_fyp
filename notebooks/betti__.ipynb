```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n_rnn = 256
bs = 512
for acc in [95]: 
    betti_topology_list = []
    h1_means_stim1 = []
    h1_stds_stim1 = []
    h1_means_stim2 = []
    h1_stds_stim2 = []
    for seed in range(1,21):
        model_folder = f'/Volumes/Seagate Backup Plus Drive/fyp/results/2_stim_batch_size_{bs}/n_hidden_{n_rnn}/2_stim_batch_size_{bs}_n_hidden_{n_rnn}_acc_{acc}_seed_{seed}_with_noise/'
        print('model folder: ' + model_folder)
        isomap_folder = model_folder+'/isomap'
        if os.path.exists(isomap_folder):
            betti_folder = isomap_folder+'/betti'
            h1_shapes = []
            h1_maxs = []
            for num in range(128):
                betti_subspace_folder = betti_folder+f'/stim2_{num}'
                with open(betti_subspace_folder+f'/h1_lens.npy', 'rb') as f:
                    h1=np.load(f)
                    h1_max = h1.max() if h1.shape[0]>0 else 0
                    h1_maxs.append(h1_max)
                    h1=[h for h in h1 if h>3]
#                     
                h1_shapes.append(len(h1))
            stim1_is_ring = (np.array(h1_shapes)!=0).any()
            print(h1_shapes)
            h1_means_stim1.append(np.array(h1_maxs).mean())
            h1_stds_stim1.append(np.array(h1_maxs).std())

            
            h1_shapes = []
            h1_maxs = []
            for num in range(128):
                betti_subspace_folder = betti_folder+f'/stim1_{num}'
                with open(betti_subspace_folder+f'/h1_lens.npy', 'rb') as f:
                    h1=np.load(f)
                    h1_max = h1.max() if h1.shape[0]>0 else 0
                    h1_maxs.append(h1_max)
                    h1=[h for h in h1 if h>3]
                    
                h1_shapes.append(len(h1))
            stim2_is_ring = (np.array(h1_shapes)!=0).any()
            print(h1_shapes)
            h1_means_stim2.append(np.array(h1_maxs).mean())
            h1_stds_stim2.append(np.array(h1_maxs).std())
            
            
            print('stim1_is_ring','stim2_is_ring','torus')
            print(stim1_is_ring,stim2_is_ring,stim1_is_ring and stim2_is_ring)
            
            betti_topology_list.append([stim1_is_ring,stim2_is_ring,stim1_is_ring and stim2_is_ring])
betti_topology_df = pd.DataFrame(betti_topology_list,columns=['stim1_is_ring','stim2_is_ring','torus'])
betti_topology_df.index = np.arange(1,21)
betti_topology_df
```

```python
stim1_is_ring = np.array([h1_means_stim1[i]-h1_stds_stim1[i]>0 for i in range(20)])
stim2_is_ring = np.array([h1_means_stim2[i]-h1_stds_stim2[i]>0 for i in range(20)])
torus = np.array([stim1_is_ring[i] and stim2_is_ring[i] for i in range(20)])
betti_topology_df_ = pd.DataFrame(np.array([stim1_is_ring,stim2_is_ring,torus]).T,columns=['$S_1$ Ring','$S_2$ Ring','Torus'])
betti_topology_df_.index = np.arange(1,21)
betti_topology_df_
```

```python
fig, ax = plt.subplots(1,1, figsize=(14,7))
x_pos = np.arange(20)
x_labels = np.arange(1,21)
width=0.3
ax.bar(x_pos - width/2, h1_means_stim1, yerr = h1_stds_stim1, width=width, align='center', alpha=0.9, ecolor='black', capsize=5, label='$S_{1}$ ring')
ax.bar(x_pos + width/2, h1_means_stim2, yerr = h1_stds_stim2, width=width, align='center', alpha=0.9, ecolor='black', capsize=5, label='$S_{2}$ ring')
ax.set_ylabel('Ring Homology (1st Betti Number)', fontsize=16)
ax.set_xlabel('Experiments', fontsize=16)
ax.set_xticks(x_pos)
ax.set_xticklabels(x_labels, fontsize=14)
plt.yticks(fontsize=14)
ax.legend(loc='upper left', bbox_to_anchor= (1, 1.02),fontsize=14)

fig.suptitle('Number of hidden neurons = 256', fontsize=20,y=0.95)
```
