```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
for acc in [60,80,95]:
    df_summary = pd.read_csv(f'/Users/lowxizhi/Documents/fyp/codes/results/2_stim_batch_size_512/n_hidden_256/summary_acc_{acc}.csv', index_col=0)

    x_pos = np.arange(len(df_summary))
    width = 0.4
    x_labels = df_summary.index

    y1_sep = df_summary['delay 1 separability mean']
    yerr1_sep = df_summary['delay 1 separability std']
    y2_sep = df_summary['delay 2 separability mean']
    yerr2_sep = df_summary['delay 2 separability std']

    y1_sel = df_summary['delay 1 selectivity mean']
    yerr1_sel = df_summary['delay 1 selectivity std']
    y2_sel = df_summary['delay 2 selectivity mean']
    yerr2_sel = df_summary['delay 2 selectivity std']

    fig, ax = plt.subplots(1,2, figsize=(16,5))
    ax[0].bar(x_pos - width/2, y1_sep, yerr = yerr1_sep, width=width, align='center', alpha=0.9, ecolor='black', capsize=5, label='Separability Index')
    ax[0].bar(x_pos + width/2, y1_sel, yerr = yerr1_sel, width=width, align='center', alpha=0.9, ecolor='black', capsize=5, label='Selectivity Index')
    ax[0].set_xlabel('Experiments')
    ax[0].set_xticks(x_pos)
    ax[0].set_xticklabels(x_labels)
    ax[0].set_title('Delay 1')
    ax[0].yaxis.grid(True)

    ax[1].bar(x_pos - width/2, y2_sep, yerr = yerr2_sep, width=width, align='center', alpha=0.9, ecolor='black', capsize=5, label='Separability Index')
    ax[1].bar(x_pos + width/2, y2_sel, yerr = yerr2_sel, width=width, align='center', alpha=0.9, ecolor='black', capsize=5, label='Selectivity Index')
    ax[1].set_xlabel('Experiments')
    ax[1].set_xticks(x_pos)
    ax[1].set_xticklabels(x_labels)
    ax[1].set_title('Delay 2')
    ax[1].yaxis.grid(True)
    ax[1].legend(loc='upper left', bbox_to_anchor= (1.1, 1.03))

    # Save the figure and show
    plt.tight_layout()
    # plt.savefig('bar_plot_with_error_bars.png')
    plt.show()
```

```python
fig, ax = plt.subplots(3,2, figsize=(15,15))
width = 0.4
x_labels = ['NS','CS1','CS2','LMS','NLMS']
x_pos = np.arange(len(x_labels))

for j, n_rnn in enumerate([128,256]):
    for i, acc in enumerate([60,80,95]):
        df_summary = pd.read_csv(f'/Users/lowxizhi/Documents/fyp/codes/results/2_stim_batch_size_512/n_hidden_{n_rnn}/summary_sel_and_sep_df_delay_2_acc_{acc}_corrected.csv', index_col=0, nrows=20)
        df_summary_0 = df_summary[df_summary['torus']==0]
        df_summary_1 = df_summary[df_summary['torus']==1]
        mean_0 = df_summary_0.iloc[:,5:10].mean(axis=0)
        std_0 = df_summary_0.iloc[:,5:10].std(axis=0)
        mean_1 = df_summary_1.iloc[:,5:10].mean(axis=0)
        std_1 = df_summary_1.iloc[:,5:10].mean(axis=0)
        ax[i,j].bar(x_pos - width/2, mean_0, yerr = std_0, width=width, align='center', alpha=0.9, ecolor='black', capsize=5, label='non torus')
        ax[i,j].bar(x_pos + width/2, mean_1, yerr = std_1, width=width, align='center', alpha=0.9, ecolor='black', capsize=5, label='torus')
        ax[i,j].set_title(f'acc = {acc}')
        ax[i,j].yaxis.grid(True)
        ax[i,j].set_xticks(x_pos)
        ax[i,j].set_xticklabels(x_labels)

        if i == 0:
            ax[i,j].text(0.5,1.2,f'number of hidden neurons = {n_rnn}',horizontalalignment='center',transform=ax[i,j].transAxes, fontsize = 20)
ax[0,1].legend(loc='upper left', bbox_to_anchor= (1.1, 1.03))
```

### Comparing selectivity types for networks with different n_rnn and networks with different performances

- n_rnn = 256 has more NLMS neurons and less LMS 
- % NLMS neurons decrease with performance
- % LMS neurons increase with performance

```python
width = 0.2
x_labels = ['NS','CS1','CS2','LMS','NLMS']
x_pos = np.arange(len(x_labels))

fig, ax = plt.subplots(1,2, figsize=(15,5), sharey=True)
for j, n_rnn in enumerate([128,256]):
    for i, acc in enumerate([60,80,95]):
        df_summary = pd.read_csv(f'/Users/lowxizhi/Documents/fyp/codes/results/2_stim_batch_size_512/n_hidden_{n_rnn}/summary_sel_and_sep_df_delay_2_acc_{acc}_corrected.csv', index_col=0, nrows=20)
        mean = df_summary.iloc[:,5:10].mean(axis=0)
        std = df_summary.iloc[:,5:10].std(axis=0)
        ax[j].bar(x_pos+ (i-1)*width, mean, yerr = std, width=width, align='center', alpha=0.9, ecolor='black', capsize=5, label=f'acc: {acc}')
    ax[j].yaxis.grid(True)
    ax[j].set_xticks(x_pos)
    ax[j].set_xticklabels(x_labels,fontsize=15)
    ax[j].set_title(f'number of hidden neurons = {n_rnn}', fontsize = 20)
ax[0].set_ylabel('% neurons',fontsize=15)
ax[1].legend(loc='upper left', bbox_to_anchor= (1.1, 1.02))
```

### Comparing selectivity types for non torus and torus networks

- for n_rnn=128, torus networks have more LMS neurons and less NLMS neurons

```python
fig, ax = plt.subplots(3,2, figsize=(15,15), sharey=True)
width = 0.4
x_labels = ['NS','CS1','CS2','LMS','NLMS']
x_pos = np.arange(len(x_labels))

for j, n_rnn in enumerate([128,256]):
    for i, acc in enumerate([60,80,95]):
        df_summary = pd.read_csv(f'/Users/lowxizhi/Documents/fyp/codes/results/2_stim_batch_size_512/n_hidden_{n_rnn}/summary_sel_and_sep_df_delay_2_acc_{acc}_corrected.csv', index_col=0, nrows=20)

        if j == 0:
            df_summary_0 = df_summary[df_summary['torus']==0]
            df_summary_1 = df_summary[df_summary['torus']==1]
        elif j == 1:
            df_summary_0 = df_summary[df_summary['torus gap']==0]
            df_summary_1 = df_summary[df_summary['torus gap']==1]
        mean_0 = df_summary_0.iloc[:,5:10].mean(axis=0)
        std_0 = df_summary_0.iloc[:,5:10].std(axis=0)
        mean_1 = df_summary_1.iloc[:,5:10].mean(axis=0)
        std_1 = df_summary_1.iloc[:,5:10].mean(axis=0)
        ax[i,j].bar(x_pos - width/2, mean_0, yerr = std_0, width=width, align='center', alpha=0.9, ecolor='black', capsize=5, label='non torus')
        ax[i,j].bar(x_pos + width/2, mean_1, yerr = std_1, width=width, align='center', alpha=0.9, ecolor='black', capsize=5, label='torus')
        ax[i,j].set_title(f'acc = {acc}', fontsize = 20)
        ax[i,j].yaxis.grid(True)
        ax[i,j].set_xticks(x_pos)
        ax[i,j].set_xticklabels(x_labels,fontsize=15)

        if j == 0:
            ax[i,j].set_ylabel('% neurons',fontsize=15)
        if i == 0:
            ax[i,j].text(0.5,1.2,f'number of hidden neurons = {n_rnn}',horizontalalignment='center',transform=ax[i,j].transAxes, fontsize = 20)
ax[0,1].legend(loc='upper left', bbox_to_anchor= (1.1, 1.03))
```

```python
y_labels = ['NS','CS1','CS2','LMS','NLMS']

for seed in range(19,20):

    fig,ax = plt.subplots(5,3, figsize=(15,30), sharey=True)
    for j, acc in enumerate([60,80,95]):
        df = pd.read_csv(f'/Users/lowxizhi/Documents/fyp/codes/results/2_stim_batch_size_512/n_hidden_{n_rnn}/2_stim_batch_size_512_n_hidden_{n_rnn}_acc_{acc}_seed_{seed}_with_noise/selectivity/selectivity_and_separabiltiy_of_hidden_neurons_delay_2.csv', index_col=0)
#         min_ = min(df['separability_128locations'])
#         max_ = max(df['separability_128locations'])
        bins = np.arange(0.7,1,0.01)
        x_pos = [round(i,2) for i in np.arange(min_,max_+0.01,0.1)]

        for i, col in enumerate(df.columns[:5]):
            df_0 = df[df[col]==0]
            df_1 = df[df[col]==1]
            ax[i,j].hist(df_0['separability_128locations'],alpha=0.8,bins=bins,label=f'Not {y_labels[i]} ({len(df_0)} neurons)')
            ax[i,j].hist(df_1['separability_128locations'],alpha=0.6,bins=bins,label=f'{y_labels[i]} ({len(df_1)} neurons)')
            ax[i,j].hist(df['separability_128locations'],alpha=0.2,bins=bins,label=f'All',color='black')
            ax[i,j].legend(loc='upper left')


            ax[i,j].set_xticks(x_pos)
            ax[i,j].set_xticklabels(x_pos)
            ax[i,j].set_xlabel('separabiltiy index',fontsize=15)
            if i == 0:
                ax[i,j].text(0.5,1.1,f'acc = {acc}',horizontalalignment='center',transform=ax[i,j].transAxes, fontsize = 20)
            if j == 0:
                ax[i,j].set_ylabel(y_labels[i],fontsize=15)
```

```python
x_labels = ['NS','CS1','CS2','LMS','NLMS']
x_pos = np.arange(len(x_labels))

for seed in range(1,20):
    fig,ax = plt.subplots(3,2, figsize=(15,15), sharey=True)
    x_pos = np.arange(5)
    for j, n_rnn in enumerate([128,256]):
        for i, acc in enumerate([60,80,95]):
            model_folder = f'/Users/lowxizhi/Documents/fyp/codes/results/2_stim_batch_size_512/n_hidden_{n_rnn}/2_stim_batch_size_512_n_hidden_{n_rnn}_acc_{acc}_seed_{seed}_with_noise'
            df = pd.read_csv(model_folder+'/selectivity/selectivity_and_separabiltiy_of_hidden_neurons_delay_2.csv', index_col=0)

            mean_0 = []
            mean_1 = []
            std_0 = []
            std_1 = []
            for col in df.columns[:5]:
                df_0 = df[df[col]==0]
                df_1 = df[df[col]==1]
                mean_0.append(df_0['separability_128locations'].mean())
                mean_1.append(df_1['separability_128locations'].mean())
                std_0.append(df_0['separability_128locations'].std())
                std_1.append(df_1['separability_128locations'].std())
    
            ax[i,j].bar(x_pos - width/2, mean_0, yerr = std_0, width=width, align='center', alpha=0.9, ecolor='black', capsize=5, label='no')
            ax[i,j].bar(x_pos + width/2, mean_1, yerr = std_1, width=width, align='center', alpha=0.9, ecolor='black', capsize=5, label='yes')
            ax[i,j].set_title(f'acc = {acc}')
            ax[i,j].yaxis.grid(True)

            ax[i,j].set_xticks(x_pos)
            ax[i,j].set_xticklabels(x_labels)

            if i == 0:
                ax[i,j].text(0.5,1.2,f'number of hidden neurons = {n_rnn}',horizontalalignment='center',transform=ax[i,j].transAxes, fontsize = 20)
    ax[0,1].legend(loc='upper left', bbox_to_anchor= (1.1, 1.03))
```
